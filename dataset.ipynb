```python
import pandas as pd

data = pd.read_csv('HH_Provider_Oct2023.csv')

# data = data.drop(['Provider ID', 'Hospital Name', 'Address', 'City', 'State', 'ZIP Code', 'County Name', 'Phone Number', 'Hospital Type', 'Hospital Ownership', 'Emergency Services', 'Meets criteria for meaningful use of EHRs', 'Hospital overall rating', 'Mortality national comparison', 'Safety of care national comparison', 'Readmission national comparison', 'Patient experience national comparison', 'Effectiveness of care national comparison', 'Timeliness of care national comparison', 'Efficient use of medical imaging national comparison'], axis=1)

print("Data before preprocessing:")
print(data)

data.replace('-', 0, inplace=True)

# scoate stringurile din coloanele numerice
numeric_columns = data.select_dtypes(include='number').columns
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# nu avem duplicate, dar daca am fi avut, le-am fi eliminat cu:
data.drop_duplicates(inplace=True)


print("\nData after preprocessing:")
print(data.head())
```
